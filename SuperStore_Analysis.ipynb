```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('superstore_data.csv')

# Convert 'Order_Date' to datetime
df['Order_Date'] = pd.to_datetime(df['Order_Date'], dayfirst=True)

# Set 'Order_Date' as the index
df.set_index('Order_Date', inplace=True)

# Resample the data to monthly frequency
df_monthly = df.resample('ME').sum()

# Convert datetime index to numerical format for polyfit
df_monthly['Date_Num'] = df_monthly.index.astype(np.int64) // 10**9  # Convert to seconds

# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df_monthly.index, df_monthly['Sales'], color='blue', label='Monthly Sales')

# Calculate the trend line
z = np.polyfit(df_monthly['Date_Num'], df_monthly['Sales'], 1)
p = np.poly1d(z)

# Plot the trend line
plt.plot(df_monthly.index, p(df_monthly['Date_Num']), color='red', label='Trend Line')

# Add titles and labels
plt.title('Monthly Sales with Trend Line')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.show()
```

```python
df = pd.read_csv ('superstore_data.csv')

df['Order_Date'] = pd.to_datetime(df['Order_Date'], dayfirst=True)


# Set 'Order_Date' as the index
df.set_index('Order_Date', inplace=True)

# Resample the data to monthly frequency
df_monthly = df.resample('ME').sum()

# Split into training (2015-2017) and test (2018) sets
train = df_monthly[:'2017-12']
test = df_monthly['2018-01':'2018-12']


from statsmodels.tsa.holtwinters import ExponentialSmoothing

# Fit the Holt-Winters model
hw_model = ExponentialSmoothing(train['Sales'], seasonal='add', seasonal_periods=12, initialization_method='estimated').fit()

# Forecast the sales for 2018
forecast = hw_model.forecast(steps=len(test))

from sklearn.metrics import mean_absolute_error, mean_squared_error

def print_metrics(y_true, y_pred):
    mae = mean_absolute_error(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    print(f"MAE: {mae:.2f}, MSE: {mse:.2f}, RMSE: {rmse:.2f}")

print_metrics(test['Sales'], forecast)


import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(train['Sales'], label='Train')
plt.plot(test['Sales'], label='Test')
plt.plot(forecast, label='Forecast', color='red')
plt.title('Holt-Winters Sales Forecast')
plt.xlabel('Order Date')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Load and preprocess data
df = pd.read_csv('superstore_data.csv')
df['Order_Date'] = pd.to_datetime(df['Order_Date'], dayfirst=True)
df.set_index('Order_Date', inplace=True)
df_monthly = df.resample('ME').sum()
train = df_monthly[:'2017-12']
test = df_monthly['2018-01':'2018-12']

# Holt-Winters Model
hw_model = ExponentialSmoothing(
    train['Sales'],
    trend='add',  
    seasonal='add',  
    seasonal_periods=12
).fit()
hw_forecast = hw_model.forecast(steps=len(test))

# ARIMA Model
arima_model = ARIMA(train['Sales'], order=(5, 1, 0)).fit()  
arima_forecast = arima_model.forecast(steps=len(test))



# Calculate Metrics
def print_metrics(y_true, y_pred, label):
    mae = mean_absolute_error(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    print(f"{label} - MAE: {mae:.2f}, MSE: {mse:.2f}, RMSE: {rmse:.2f}")

print_metrics(test['Sales'], hw_forecast, 'Holt-Winters')
print_metrics(test['Sales'], arima_forecast, 'ARIMA')

# Plot Results
plt.figure(figsize=(12, 8))
plt.plot(train['Sales'], label='Train')
plt.plot(test['Sales'], label='Test')
plt.plot(hw_forecast, label='Holt-Winters Forecast', color='red')
plt.plot(arima_forecast, label='ARIMA Forecast', color='green')
plt.title('Sales Forecast Comparison')
plt.xlabel('Order Date')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.holtwinters import ExponentialSmoothing

# Load and preprocess data
df = pd.read_csv('superstore_data.csv')
df['Order_Date'] = pd.to_datetime(df['Order_Date'], dayfirst=True)
df.set_index('Order_Date', inplace=True)
df_monthly = df.resample('ME').sum()

# Use data from 2015 to 2018 for training
train = df_monthly[:'2018-12']

# Holt-Winters Model
hw_model = ExponentialSmoothing(
    train['Sales'],
    trend='add', 
    seasonal='add', 
    seasonal_periods=12
).fit()

# Forecast sales for 2019 (12 months)
hw_forecast = hw_model.forecast(steps=12)

# Create a date range for the forecast
forecast_index = pd.date_range(start='2019-01-01', periods=12, freq='ME')

# Convert forecast to a DataFrame
forecast_df = pd.DataFrame(hw_forecast, index=forecast_index, columns=['Forecasted Sales'])

# Display the forecasted sales for 2019
print(forecast_df)

# Plot Results
plt.figure(figsize=(12, 8))
plt.plot(train['Sales'], label='Train', color='blue')
plt.plot(forecast_df['Forecasted Sales'], label='Holt-Winters Forecast', color='red')
plt.title('Sales Forecast for 2019')
plt.xlabel('Order Date')
plt.ylabel('Sales')
plt.legend()
plt.grid()
plt.show()
```
